```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
#from sklearn.linear_model import LogisticRegression
```

```python
mat_students = pd.read_csv('student-mat.csv', delimiter = ';')
print(mat_students.head())
```

```python
por_students = pd.read_csv('student-mat.csv', delimiter = ';')
print(por_students.head())
```

```python
mat_students.info()
```

```python
por_students.info()
```

```python
students = pd.merge(mat_students, por_students, how = 'outer', on =["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"])
```

```python
students.head()
```

```python
students.info()
```

```python
label_1 = 'G3_x'
label_2 = 'G3_y'
```

```python
plt.figure(figsize = (60, 60))
sns.set(font_scale = 2)
ax = sns.heatmap(students.corr(), annot = True, linewidth = 0.5, cmap = 'coolwarm', cbar_kws = {'label': 'Feature Correlation'})

#highlighting important features
for lab, annot in zip(ax.get_yticklabels(), ax.texts):
    text = lab.get_text()
    if text == label_1 or text == label_2:
        lab.set_weight('bold')
        lab.set_size(50)
        lab.set_color('purple')
        
       
```
